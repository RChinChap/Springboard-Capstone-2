## Preparing Traing and Testing Data

```python
#dividing data
X = df.drop("diagnosis",axis = 1)
y = df.diagnosis
```

```python
df.head(10)
```

```python
#importing train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42,shuffle=True, stratify=y)
```

```python
print('X_train shape : ', X_train.shape )
print('y_train shape : ', y_train.shape )
print('X_test shape : ', X_test.shape )
print('y_test shape : ', y_test.shape )
```

```python
from sklearn.neighbors import KNeighborsClassifier


test_scores = []
train_scores = []
K=[]
for i in range(1,15):

    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
    K.append(i)
```

```python
results= { 'train_scores':train_scores,
         'test_scores':test_scores,
         'K':K}
resultsdf =  pd.DataFrame(results,
        columns=['train_scores','test_scores', 'K'])
resultsdf
```

```python
#appling KNN model for k=1
KNNModel = KNeighborsClassifier(1)
KNNModel.fit(X_train,y_train)

train_score = KNNModel.score(X_train,y_train)
test_score = KNNModel.score(X_test,y_test)
print( 'train_score ', train_score)
print( 'test_score ',test_score )
```

```python
#prediction step
from sklearn.metrics import confusion_matrix

y_pred = knn.predict(X_test)


confusion_matrix(y_test,y_pred)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```
